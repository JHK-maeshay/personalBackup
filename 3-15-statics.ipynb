6-3. 통계 함수
통계 함수를 통해 array의 합이나 평균등을 구할 때,

추가로 axis라는 인자에 대한 값을 지정하여 열 또는 행의 합 또는 평균등을 구할 수 있다.

```python
import numpy as np
```

```python
#랜덤함수로 5x3의 array 생성
arr1 = np.random.randn(5,3)
arr1
```

```python
# 전체 성분의 합을 계산
np.sum(arr1)
```

+ 행렬은 2차원 텐서
+ [행, 열]의 axis는 각각 [0, 1]
+ 헷갈리면 행 → 0 열 ↓1

```python
# 열 간의 합을 계산
np.sum(arr1, axis=1)
```

```python
# 행 간의 합을 계산
np.sum(arr1, axis=0)
```

```python
# 전체 성분의 평균을 계산
np.mean(arr1)
```

```python
# 행 간의 평균을 계산
np.mean(arr1, axis=0)
```

```python
# 전체 성분의 표준편차, 분산, 최소값, 최대값 계산(std, var, min, max)
np.std(arr1), np.var(arr1)
```

예제

어느 학생 집단에서

+ 주당 도서관에 있는 시간을 x1
+ 패턴인식 과목 점수를 x2
+  주당 강의 빼먹는 시간을 x3

이라 하고 다음과 같은 데이터를 수집하였다.


-> (x1, x2), (x2, x3)에 대한 상관계수를 알아보아라.

arr2 = np.array(
  [35, 11, 0.5],
  [5, 12, 0],
  [0, 5, 10],
  [20, 9, 4.5],
  [15, 8, 5])

  -> 평균, 분산, 공분산, 상관계수

```python
# 데이터
arr2 = np.array([
  [35, 11, 0.5],
  [5, 12, 0],
  [0, 5, 10],
  [20, 9, 4.5],
  [15, 8, 5]])
# 평균
arrmean=np.mean(arr2, axis=0)
arrmean
```

```python
# 분산 (n으로 나눔)
arrvar=np.var(arr2, axis=0)
arrvar
```

```python
# 공분산
arrx1 = arr2[:,0]
arrx2 = arr2[:,1]
arrx3 = arr2[:,2]
# 두 개의 array에 대해 동일한 위치의 성분끼리 연산 값을 계산하기(add, subtract, multiply, divide)
# 공분산 (x1, x2), (x2, x3)
# (xi-xmean)의 값을 곱해야 하므로 subtract와 ones로 구현
c12=(np.sum(np.multiply(
  np.subtract(arrx1,arrmean[0]*np.ones(5)),
  np.subtract(arrx2,arrmean[1]*np.ones(5))
  )))/(5-1)
c23=(np.sum(np.multiply(
  np.subtract(arrx2,arrmean[1]*np.ones(5)),
  np.subtract(arrx3,arrmean[2]*np.ones(5))
  )))/(5-1)
c12, c23
```

```python
#sqrt를 쓰기 위해서 math 라이브러리 사용
import math
r12=c12/math.sqrt(arrvar[0])/math.sqrt(arrvar[1])
r23=c23/math.sqrt(arrvar[1])/math.sqrt(arrvar[2])
r12, r23
```

문제의 답은 0.4667, -0.9926이지만,

numpy 라이브러리 함수 variance가 n-1이 아니라 n으로 나누는 차이가 있으므로

계산방법은 적절했다고 볼 수 있다.

```python
arr3 = np.array([
  [77, 85, 48, 94, 80, 60, 65, 96, 70, 86, 69, 82],
  [78.2, 82.1, 60.3, 91.5, 84.6, 70.8, 68.4, 90.8, 75.0, 92.5, 61.8, 80.1]])
# 평균
arrmean=np.mean(arr3, axis=1)


# 분산 (n으로 나눔)
arrvar=np.var(arr3, axis=1)
arrmean, arrvar
```

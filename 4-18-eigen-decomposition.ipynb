1. **공분산행렬의 특성**

* A. symmetric 하다 ($A^{-1}=A^T)$

+ B. 고유 벡터가 orthogonal

+ C. eigenvalue가 양수다 $(λ > 0)$

2. **고유값 분해**
***

행렬 $A$에 대해...

(1) $Av = λv$

대각성분이 $λ_i$인 대각행렬을 $D$라 하자..

또, $v_i$를 열벡터로 하는 행렬을 $U$라 하자..

(단, $|v_i|=1$이다)

(2) $AU = UD$

따라서...

(3) $A=UDU^{-1} \text{ and }U^{-1}AU=D$

A가 symmetric한 공분산행렬 C라고 한다면...

(4) $C = UDU^T$

```python
def cova(data):
  print("*input matrix <class = %d>" % (len(data.T)))
  print(data)
  datamean = np.mean(data, axis=0)
  datamu = np.subtract(data, datamean)
  dataOut = np.multiply(np.dot(datamu.T, datamu), 1/(len(data)-1))
  print("==> covariance matrix")
  print(dataOut)
  return dataOut
```

```python
# 데이터
arr2 = np.array([
  [35, 11, 0.5],
  [5, 12, 0],
  [0, 5, 10],
  [20, 9, 4.5],
  [15, 8, 5]])
# 평균
arrmean=np.mean(arr2, axis=0)
arrmean
```

```python
# 분산 (n으로 나눔)
arrvar=np.var(arr2, axis=0)
arrvar
```

```python
# 공분산
arrx1 = arr2[:,0]
arrx2 = arr2[:,1]
arrx3 = arr2[:,2]
# 두 개의 array에 대해 동일한 위치의 성분끼리 연산 값을 계산하기(add, subtract, multiply, divide)
# 공분산 (x1, x2), (x2, x3)
# (xi-xmean)의 값을 곱해야 하므로 subtract와 ones로 구현
c12=(np.sum(np.multiply(
  np.subtract(arrx1,arrmean[0]*np.ones(5)),
  np.subtract(arrx2,arrmean[1]*np.ones(5))
  )))/(5-1)
c23=(np.sum(np.multiply(
  np.subtract(arrx2,arrmean[1]*np.ones(5)),
  np.subtract(arrx3,arrmean[2]*np.ones(5))
  )))/(5-1)
c12, c23
```
